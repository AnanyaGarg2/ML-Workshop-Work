```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([[1, 0], [2, 0], [3, 0], [4, 0]])
y = np.array([1, 4, 9, 16])

# Polynomial features
poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(x)

# Model fitting
model = LinearRegression()
model.fit(poly_features, y)

# Predictions
y_pred = model.predict(poly_features)

# Plotting
plt.title("POLY REGRESSION MODEL")
plt.scatter(x[:, 0], y, color='blue', label='Actual')
plt.plot(x[:, 0], y_pred, color="black", label='Prediction')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([[2, 0], [5, 8], [8, 0], [9, 3]])
y = np.array([1, 4, 9, 16])

# Polynomial features
poly = PolynomialFeatures(degree=2)
poly_features = poly.fit_transform(x)

# Model fitting
model = LinearRegression()
model.fit(poly_features, y)

# Predictions
y_pred = model.predict(poly_features)

# Plotting
plt.title("POLY REGRESSION MODEL")
plt.scatter(x[:, 0], y, color='blue', label='Actual')
plt.plot(x[:, 0], y_pred, color="black", label='Prediction')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
x = np.array([[1, 0], [2, 0], [3, 0], [4, 0]])
y = np.array([1, 4, 9, 16])

# Polynomial features
poly = PolynomialFeatures(degree=5)
poly_features = poly.fit_transform(x)

# Model fitting
model = LinearRegression()
model.fit(poly_features, y)

# Predictions
y_pred = model.predict(poly_features)

# Plotting
plt.title("POLY REGRESSION MODEL")
plt.scatter(x[:, 0], y, color='blue', label='Actual')
plt.plot(x[:, 0], y_pred, color="black", label='Prediction')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
